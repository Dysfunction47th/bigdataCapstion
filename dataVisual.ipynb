```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 데이터 프레임 생성
sample_1 = pd.read_csv('./경찰청_졸음운전 교통사고 현황_20191231.csv', encoding='cp949')

df = pd.DataFrame(sample_1)

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/gulim.ttc"  # 설치한 한글 폰트의 경로를 입력해야 합니다
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# Unique road types
road_types = df['도로종류'].unique()

# Create separate subplots for each road type
fig, axs = plt.subplots(len(road_types), figsize=(12, 6 * len(road_types)))

for i, road_type in enumerate(road_types):
    temp_df = df[df['도로종류'] == road_type]
    ax = axs[i]
    bar_width = 0.2
    index = range(len(temp_df['구분'].unique()))

    # Bar plots for '사고(건)', '사망(명)', '부상(명)'
    ax.bar([x for x in index], temp_df['사고(건)'], bar_width, label='사고(건)')
    ax.bar([x + bar_width for x in index], temp_df['사망(명)'], bar_width, label='사망(명)')
    ax.bar([x + 2 * bar_width for x in index], temp_df['부상(명)'], bar_width, label='부상(명)')


    ax.set_title(f'도로종류: {road_type}')
    ax.set_xlabel('연도')
    ax.set_ylabel('건수')
    ax.legend()
    ax.grid(True)

    # Update the x-axis tick labels to match the '구분' values
    ax.set_xticks([x + 3 * bar_width for x in index])
    ax.set_xticklabels(temp_df['구분'].unique())

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

sample_1 = pd.read_csv('경찰청_차종별 졸음운전 교통사고 현황_12_31_2020.csv', encoding='cp949')
sample_11 = sample_1[sample_1['구분'] == '부상(명)']

vehicle_types = ['승용차', '승합차', '이륜차', '특수차', '화물차']
years = sample_11['년도']

plt.figure(figsize=(12, 6))

bar_width = 0.15  # Width of the bars
index = range(len(years))

for i, vehicle_type in enumerate(vehicle_types):
    plt.bar([x + i * bar_width for x in index], sample_11[vehicle_type], bar_width, label=vehicle_type)

plt.title('차종별 졸음운전 교통사고 현황')
plt.xlabel('년도')
plt.ylabel('부상(명)')
plt.xticks([x + (bar_width * len(vehicle_types)) / 2 for x in index], years)
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 데이터 프레임 생성
sample_1 = pd.read_csv('./경찰청_졸음운전 교통사고 현황_20191231.csv', encoding='cp949')

df = pd.DataFrame(sample_1)

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/gulim.ttc"  # 설치한 한글 폰트의 경로를 입력해야 합니다
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# Filter data for '일반국도' only
filtered_data = df[df['도로종류'] == '일반국도']

# Create a single subplot
fig, ax = plt.subplots(figsize=(12, 6))

bar_width = 0.2
index = range(len(filtered_data['구분'].unique()))

# Bar plots for '사고(건)', '사망(명)', '부상(명)'
ax.bar([x for x in index], filtered_data['사고(건)'], bar_width, label='사고(건)')
ax.bar([x + bar_width for x in index], filtered_data['사망(명)'], bar_width, label='사망(명)')
ax.bar([x + 2 * bar_width for x in index], filtered_data['부상(명)'], bar_width, label='부상(명)')

ax.set_title('일반국도의 졸음운전 교통사고 현황')
ax.set_xlabel('연도')
ax.set_ylabel('건수')
ax.legend()
ax.grid(True)

# Update the x-axis tick labels to match the '구분' values
ax.set_xticks([x + 3 * bar_width for x in index])
ax.set_xticklabels(filtered_data['구분'].unique())

plt.tight_layout()
plt.show()
```
